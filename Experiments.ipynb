## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1: Salary vs Gender


#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- Null Hypothesis (Hâ‚€): There is no significant difference in monthly salary between male and female employees.

- Alternate Hypothesis (Hâ‚): There is a significant difference in monthly salary between male and female employees.

#### 2. Perform an appropriate statistical test.

````python
from scipy import stats
````

````python
# Perform Statistical Test to obtain P-Value
male_salary = df[df['gender'] == 'Male']['monthly_salary']
female_salary = df[df['gender'] == 'Female']['monthly_salary']

t_stat, p_val = stats.ttest_ind(male_salary, female_salary, equal_var=False)
print("Test 1: Salary vs Gender")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Reject Hâ‚€ â†’ Salary differs significantly between genders\n")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference in salary by gender\n")
````

##### Which statistical test have you done to obtain P-Value?

Independent Samples t-test (Studentâ€™s t-test)

##### Why did you choose the specific statistical test?

Because we are comparing the means of two independent groups (Male vs Female) for a continuous variable (Monthly Salary).

###  Hypothetical Statement - 2: Credit Score vs EMI Eligibility

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- Null Hypothesis (Hâ‚€): Credit score and EMI eligibility are independent.

- Alternate Hypothesis (Hâ‚): Credit score and EMI eligibility are dependent (related).

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
contingency = pd.crosstab(df['emi_eligibility'], pd.cut(df['credit_score'], bins=5))
chi2, p_val, dof, expected = stats.chi2_contingency(contingency)
print("Test 2: Credit Score vs EMI Eligibility")
print(f"Chi2: {chi2:.4f}, P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Reject Hâ‚€ â†’ Credit score and EMI eligibility are related\n")
else:
    print("Fail to reject Hâ‚€ â†’ No significant relationship\n")
````

##### Which statistical test have you done to obtain P-Value?

Chi-Square Test of Independence

##### Why did you choose the specific statistical test?

Because both variables (Credit Score Category and EMI Eligibility) are categorical, and we want to check for association between them.

### 2. Handling Outliers

````python
# Enhanced LogisticRegression Model Training & Evaluation with MLflow Logging
import mlflow
import mlflow.sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    precision_score, recall_score, f1_score, roc_auc_score,
    roc_curve, precision_recall_curve
)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import cross_val_score

with mlflow.start_run(run_name="LogisticRegression_Enhanced"):
    # Define model with explicit hyperparameters
    log_reg = LogisticRegression(
        max_iter=1000,
        random_state=42,
        solver='lbfgs',
        penalty='l2',
        C=1.0
    )
    
    # Train model
    log_reg.fit(X_train_cls_bal, y_train_cls_bal)
    y_pred_cls = log_reg.predict(X_test_cls)
    y_pred_proba = log_reg.predict_proba(X_test_cls)
    
    # Calculate comprehensive metrics
    acc = accuracy_score(y_test_cls, y_pred_cls)
    precision = precision_score(y_test_cls, y_pred_cls, average='weighted')
    recall = recall_score(y_test_cls, y_pred_cls, average='weighted')
    f1 = f1_score(y_test_cls, y_pred_cls, average='weighted')
    
    # Cross-validation scores
    cv_scores = cross_val_score(log_reg, X_train_cls_bal, y_train_cls_bal, cv=5)
    
    # === LOG PARAMETERS ===
    mlflow.log_param("model_name", "LogisticRegression")
    mlflow.log_param("max_iter", 1000)
    mlflow.log_param("random_state", 42)
    mlflow.log_param("solver", "lbfgs")
    mlflow.log_param("penalty", "l2")
    mlflow.log_param("C", 1.0)
    mlflow.log_param("training_samples", len(X_train_cls_bal))
    mlflow.log_param("test_samples", len(X_test_cls))
    mlflow.log_param("n_features", X_train_cls_bal.shape[1])
    mlflow.log_param("n_classes", len(np.unique(y_train_cls_bal)))
    
    # === LOG METRICS ===
    mlflow.log_metric("accuracy", acc)
    mlflow.log_metric("precision", precision)
    mlflow.log_metric("recall", recall)
    mlflow.log_metric("f1_score", f1)
    mlflow.log_metric("cv_mean_score", cv_scores.mean())
    mlflow.log_metric("cv_std_score", cv_scores.std())
    mlflow.log_metric("n_iterations", log_reg.n_iter_[0] if hasattr(log_reg, 'n_iter_') else 0)
    
    # ROC AUC (for binary or multi-class)
    if len(np.unique(y_test_cls)) == 2:
        roc_auc = roc_auc_score(y_test_cls, y_pred_proba[:, 1])
        mlflow.log_metric("roc_auc", roc_auc)
    else:
        roc_auc = roc_auc_score(y_test_cls, y_pred_proba, multi_class='ovr', average='weighted')
        mlflow.log_metric("roc_auc_weighted", roc_auc)
    
    # === VISUALIZATIONS ===
    
    # 1. Confusion Matrix
    cm = confusion_matrix(y_test_cls, y_pred_cls)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True)
    plt.title('Confusion Matrix - LogisticRegression')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    mlflow.log_figure(plt.gcf(), "confusion_matrix.png")
    plt.close()
    
    # 2. Coefficient Weights (Feature Importance for Logistic Regression)
    if log_reg.coef_.shape[0] == 1:  # Binary classification
        coefficients = log_reg.coef_[0]
    else:  # Multi-class - show first class or average
        coefficients = np.abs(log_reg.coef_).mean(axis=0)
    
    plt.figure(figsize=(12, 6))
    indices = np.argsort(np.abs(coefficients))[::-1][:20]  # Top 20 features
    plt.bar(range(len(indices)), np.abs(coefficients[indices]))
    plt.title('Top 20 Feature Coefficients (Absolute Values)')
    plt.xlabel('Feature Index')
    plt.ylabel('Coefficient Magnitude')
    plt.tight_layout()
    mlflow.log_figure(plt.gcf(), "feature_coefficients.png")
    plt.close()
    
    # 3. ROC Curve (for binary classification)
    if len(np.unique(y_test_cls)) == 2:
        fpr, tpr, _ = roc_curve(y_test_cls, y_pred_proba[:, 1])
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.3f})', linewidth=2)
        plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        mlflow.log_figure(plt.gcf(), "roc_curve.png")
        plt.close()
        
        # 4. Precision-Recall Curve
        precision_curve, recall_curve, _ = precision_recall_curve(y_test_cls, y_pred_proba[:, 1])
        plt.figure(figsize=(8, 6))
        plt.plot(recall_curve, precision_curve, linewidth=2)
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curve')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        mlflow.log_figure(plt.gcf(), "precision_recall_curve.png")
        plt.close()
    
    # 5. Class Distribution
    unique, counts = np.unique(y_test_cls, return_counts=True)
    plt.figure(figsize=(8, 6))
    plt.bar(unique, counts, alpha=0.7, label='True', width=0.4)
    unique_pred, counts_pred = np.unique(y_pred_cls, return_counts=True)
    plt.bar(unique_pred + 0.4, counts_pred, alpha=0.7, label='Predicted', width=0.4)
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.title('Class Distribution: True vs Predicted')
    plt.legend()
    plt.xticks(unique + 0.2, unique)
    plt.tight_layout()
    mlflow.log_figure(plt.gcf(), "class_distribution.png")
    plt.close()
    
    # 6. Cross-Validation Scores
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(cv_scores) + 1), cv_scores, marker='o', linestyle='-', linewidth=2, markersize=8)
    plt.axhline(y=cv_scores.mean(), color='r', linestyle='--', label=f'Mean: {cv_scores.mean():.3f}')
    plt.fill_between(range(1, len(cv_scores) + 1), 
                     cv_scores.mean() - cv_scores.std(), 
                     cv_scores.mean() + cv_scores.std(), 
                     alpha=0.2, color='r')
    plt.xlabel('Fold')
    plt.ylabel('Score')
    plt.title('Cross-Validation Scores')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    mlflow.log_figure(plt.gcf(), "cv_scores.png")
    plt.close()
    
    # 7. Decision Boundary Probabilities (for binary classification)
    if len(np.unique(y_test_cls)) == 2:
        plt.figure(figsize=(10, 6))
        plt.hist(y_pred_proba[y_test_cls == 0][:, 1], bins=50, alpha=0.5, label='Class 0', density=True)
        plt.hist(y_pred_proba[y_test_cls == 1][:, 1], bins=50, alpha=0.5, label='Class 1', density=True)
        plt.axvline(x=0.5, color='r', linestyle='--', label='Decision Threshold')
        plt.xlabel('Predicted Probability (Class 1)')
        plt.ylabel('Density')
        plt.title('Predicted Probability Distribution by True Class')
        plt.legend()
        plt.tight_layout()
        mlflow.log_figure(plt.gcf(), "probability_distribution.png")
        plt.close()
    
    # === LOG ARTIFACTS ===
    # Log classification report as text file
    report = classification_report(y_test_cls, y_pred_cls)
    with open("classification_report.txt", "w") as f:
        f.write(report)
    mlflow.log_artifact("classification_report.txt")
    
    # Log confusion matrix as numpy array
    np.save("confusion_matrix.npy", cm)
    mlflow.log_artifact("confusion_matrix.npy")
    
    # Log coefficients
    np.save("model_coefficients.npy", log_reg.coef_)
    mlflow.log_artifact("model_coefficients.npy")
    
    # === LOG MODEL ===
    mlflow.sklearn.log_model(
        log_reg, 
        "LogisticRegression",
        signature=mlflow.models.infer_signature(X_train_cls_bal, y_pred_cls)
    )
    
    # === PRINT RESULTS ===
    print("=== LogisticRegression - Enhanced Tracking ===")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print(f"\nCross-Validation Scores: {cv_scores}")
    print(f"Mean CV Score: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
    print(f"Number of iterations: {log_reg.n_iter_[0] if hasattr(log_reg, 'n_iter_') else 'N/A'}")
    print("\nConfusion Matrix:\n", cm)
    print("\nClassification Report:\n", report)
    print(f"\nMLflow Run ID: {mlflow.active_run().info.run_id}")
````
